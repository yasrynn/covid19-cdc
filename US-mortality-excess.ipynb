```python
df = pd.read_csv('Weekly_Counts_of_Deaths_by_State_and_Select_Causes__2019-2020.csv')
```

```python
df.columns
```

```python
agg_df = df[(df['Jurisdiction of Occurrence'] == 'United States')][['MMWR Week', 'MMWR Year', 'All Cause', 'Influenza and pneumonia (J09-J18)']]
```

```python
agg_df.head()
```

```python
cutoff_week = 34
```

```python
cause = 'All Cause'

ctrl_df = agg_df[agg_df['MMWR Year'] == 2019]
test_df = agg_df[(agg_df['MMWR Year'] == 2020) & (agg_df['MMWR Week'] <= cutoff_week)]
# test_df = agg_df[(agg_df['MMWR Year'] == 2020)]

plt.figure(figsize=(12,10))
plt.plot(ctrl_df['MMWR Week'], ctrl_df[cause], label=2019)
plt.plot(test_df['MMWR Week'], test_df[cause], label=2020)
plt.xlabel('Week in Calendar Year')
plt.ylabel(f'Number of Deaths in US by {cause}')
plt.ylim(0, 100000)
plt.legend()
```

```python
test_df[test_df['MMWR Week'] <= cutoff_week][cause].sum() - ctrl_df[ctrl_df['MMWR Week'] <= cutoff_week][cause].sum()
```

```python
cause = 'Influenza and pneumonia (J09-J18)'

ctrl_df = agg_df[agg_df['MMWR Year'] == 2019]
test_df = agg_df[(agg_df['MMWR Year'] == 2020) & (agg_df['MMWR Week'] <= cutoff_week)]

plt.figure(figsize=(12,10))
plt.plot(ctrl_df['MMWR Week'], ctrl_df[cause], label=2019)
plt.plot(test_df['MMWR Week'], test_df[cause], label=2020)
plt.xlabel('Week in Calendar Year')
plt.ylabel(f'Number of Deaths in US due to {cause}')
plt.ylim(0)
plt.legend()
```

```python
test_df[test_df['MMWR Week'] <= cutoff_week][cause].sum() - ctrl_df[ctrl_df['MMWR Week'] <= cutoff_week][cause].sum()
```
